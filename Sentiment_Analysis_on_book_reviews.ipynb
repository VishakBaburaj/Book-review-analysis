# Text pre processing

# Converting case into lower

```python
test_data["Cleaned_Comments"] = test_data["Comments"].str.lower()
test_data.head(5)
```

```python
test_data["Cleaned_Comments"] = [remove_special_characters(sentence) for sentence in test_data["Cleaned_Comments"]] 
test_data["Cleaned_Comments"]
```

# Removing Multiple Whitespaces

```python
test_data["Cleaned_Comments"] = test_data["Cleaned_Comments"].replace('\s+', ' ', regex = True)
test_data["Cleaned_Comments"]
```

# Removing Numbers

```python
test_data["Cleaned_Comments"] = test_data["Cleaned_Comments"].str.replace('\d+','')
test_data["Cleaned_Comments"]
```

# Converting Spelling Mistakes to correct spelling

```python
from autocorrect import Speller
spell = Speller(lang='en')
```

```python
test_data["Cleaned_Comments"] = test_data["Cleaned_Comments"].apply(lambda x: " ".join([spell(i) for i in x.split()]))
test_data["Cleaned_Comments"]
```

# Removing Stopwords

```python
from nltk.corpus import stopwords
STOPWORDS = stopwords.words('english')
#STOPWORDS.extend([])
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
test_data["Cleaned_Comments"] = test_data["Cleaned_Comments"].apply(lambda text: remove_stopwords(text))
test_data["Cleaned_Comments"]
```

# Sentiment Analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

neg = []
pos = []
neu = []
compound_score = []

analyzer = SentimentIntensityAnalyzer()
for cleaned_comment in test_data["Cleaned_Comments"]:
    vs = analyzer.polarity_scores(cleaned_comment)
    negative = vs['neg']
    positive = vs['pos']
    neutral = vs['neu']
    compound = vs['compound']
    neg.append(float(negative))
    pos.append(float(positive))
    neu.append(float(neutral))
    compound_score.append(compound)
```

```python
test_data['Negative'] = neg
test_data['Positive'] = pos
test_data['Neutral'] = neu
test_data['Compund_Score'] = compound_score
test_data
```

```python
test_data['sentiment'] = 'neutral_review'
test_data.loc[test_data['Compund_Score'] >= 0.05, 'sentiment'] = 'positive_review'
test_data.loc[test_data['Compund_Score'] <= -0.05, 'sentiment'] = 'negative_review'
test_data.head(5)
```
